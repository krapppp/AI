### 라이브러리 임포트

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Input
```

### 탐색적 데이터 분석 결과 탑재

```python
# 구글 드라이드
from google.colab import drive
drive.mount('/content/drive')

ROOTDIR="/content/drive/MyDrive/"
```

```python
# read churn data
df = pd.read_csv(ROOTDIR + '/DATA/processed_churn_data.csv')
```

```python
print(df.info())
```

```python
print(df.isna().sum())
```

### 더미 특성 생성

```python
cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
        'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
        'PaperlessBilling', 'PaymentMethod']
```

```python
# One-hot encoding : get_dummies
dummies = pd.get_dummies(df[cols], drop_first=True)
```

```python
# dorp columns
df = df.drop(cols, axis=1)
```

```python
# concat
df = pd.concat([df, dummies], axis=1)
```

```python
pd.options.display.max_columns = 50
pd.options.display.max_rows = 50
```

```python
df.head(3)
```

###  Train, Test  데이터셋 분할

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Churn', axis=1).values
y = df['Churn'].values
```

```python
# train_test_splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
# shape
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 데이터 정규화/스케일링(Normalizing/Scaling)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test[0]
```

```python
X_test.shape
```

###  모델 구성


```python
model = keras.Sequential()
model.add(keras.layers.Input((23,)))
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(32, activation='relu'))

model.add(keras.layers.Dense(1, activation='sigmoid'))
```

### 모델 확인

```python
model.summary()
```

#### 모델 컴파일 – 이진 분류 모델

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

- 모델 컴파일 – 다중 분류 모델  
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


- 모델 컴파일 – 예측 모델
model.compile(optimizer='adam', loss='mse')

### 모델 훈련(학습) 하기

```python
from tqdm.keras import TqdmCallback

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    batch_size=32,
    epochs=30,
    verbose=0,
    callbacks=[TqdmCallback()]
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()
```

### 평가

```python
loss, accuracy = model.evaluate(X_test, y_test)

print(f"테스트 손실: {loss:.4f}")
print(f"테스트 정확도: {accuracy:.4f}")
```

```python
from sklearn.metrics import classification_report
y_pred_proba = model.predict(X_test)
y_pred_proba
```

```python
# 확률을 0 또는 1로 변환 (임계값 0.5)
y_pred = (y_pred_proba > 0.5).astype(int)
```

```python
print(classification_report(y_test, y_pred))
```

### 모델 훈련(학습) - 조기종료

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

과적합 방지

```python
model2 = Sequential()
model2.add(Input(shape=(23,)))
# model2.add(Dense(128, activation='relu'))
# model2.add(Dropout(0.5))
model2.add(Dense(64, activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(64, activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(32, activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss',
                           mode='min',
                           verbose=0,
                           patience=25)
```

```python
check_point =  ModelCheckpoint('models/churn_best_model.h5', verbose=1,
                              monitor='val_loss', mode='min',
                              save_best_only=True)
```

```python
model2.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
history2 = model2.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    batch_size=32,
    epochs=50,
    verbose=0,
    callbacks=[TqdmCallback(), early_stop, check_point]
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)

plt.plot(history2.history['loss'], 'b-', label='loss')
plt.plot(history2.history['val_loss'], 'r--', label='val_loss')

plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history2.history['accuracy'], 'g-', label='accuracy')
plt.plot(history2.history['val_accuracy'], 'k--', label='val_accuracy')

plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()
```

```python
loss, accuracy = model2.evaluate(X_test, y_test)

print(f"테스트 손실: {loss:.4f}")
print(f"테스트 정확도: {accuracy:.4f}")
```
