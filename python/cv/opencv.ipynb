```python
import cv2
import numpy as np

# 흰 배경 이미지 생성 (600x600, 3채널)
img = np.ones((600, 600, 3), dtype=np.uint8) * 255

# 1. (10, 10)에서 (500, 500)까지 파란색 직선, 두께 2
a=cv2.line(img, (10, 10), (500, 500), color=(0, 0, 255), thickness=2)

# 2. (100, 400)에서 (400, 400)까지 초록색 수평선, 선 타입은 LINE_AA
b=cv2.line(img, (100, 400), (400, 400), color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)

# 3. (50, 50)에서 (300, 50)까지 빨간색 화살표, tipLength=0.2
c=cv2.arrowedLine(img, (50, 50), (300, 50), color=(255, 0, 0), thickness=2, tipLength=0.2)

# 4. (250, 100)에서 (250, 400)까지 아래 방향 보라색 화살표, tipLength=0.1
d=cv2.arrowedLine(img, (250, 100), (250, 400), color=(255, 0, 255), thickness=2, tipLength=0.1)

# 5. (50, 50)에서 (300, 200)까지 초록색 테두리 사각형, 두께 2
e=cv2.rectangle(img, (50, 50), (300, 200), color=(0, 255, 0), thickness=2)

# 이미지 출력
# cv2.imshow('Drawing Practice', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(a)
plt.imshow(b)
plt.imshow(c)
plt.imshow(d)
plt.imshow(e)
# plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

src1 = cv2.imread('./img/abs1.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./img/abs2.jpg', cv2.IMREAD_GRAYSCALE)
dst = cv2.absdiff(src1, src2)

plt.subplot(131), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('abs1')
plt.subplot(132), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('abs2')
plt.subplot(133), plt.axis('off'), plt.imshow(dst, 'gray'), plt.title('result')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('./img/candies.png')
b, g, r = cv2.split(image)
merge_b = cv2.merge([b, b, b])
merge_g = cv2.merge([g, g, g])
merge_r= cv2.merge([r, r, r])

plt.subplot(221), plt.axis('off'), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(222), plt.axis('off'), plt.imshow(merge_b, cmap='gray'), plt.title('Blue Channel')
plt.subplot(223), plt.axis('off'), plt.imshow(merge_g, cmap='gray'), plt.title('Green Channel')
plt.subplot(224), plt.axis('off'), plt.imshow(merge_r, cmap='gray'), plt.title('Red Channel')
plt.show()
```

```python
import cv2, sys
import numpy as np

src = cv2.imread('./img/neutrophils.png', cv2.IMREAD_GRAYSCALE)
def on_threshold(pos):
  _,dst= cv2.threshold (src, pos, 255, cv2. THRESH_BINARY)
  cv2.imshow('dst', dst)

cv2.imshow('src', src)
cv2.namedWindow('dst')
cv2.createTrackbar ('Threshold', 'dst', 0, 255, on_threshold)
cv2.setTrackbarPos ('Threshold', 'dst', 128)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2, sys
src= cv2.imread('./img/keyboard.png', cv2.IMREAD_GRAYSCALE)
_, src_bin = cv2.threshold (src, 0, 255, cv2.THRESH_OTSU)
cnt, labels, stats, centroids = cv2.connectedComponentsWithStats (src_bin)
dst = cv2.cvtColor(src, cv2.COLOR_GRAY2BGR)

for i in range(1, cnt):
  (x, y, w, h, area) = stats[i]
  if area < 20:
    continue
  cv2.rectangle(dst, (x, y, w, h), (0, 255, 255))

cv2.imshow('src', src)
cv2.imshow('src_bin', src_bin)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2, random
src = cv2.imread('./img/contours.png', cv2.IMREAD_GRAYSCALE)
contours, hier=cv2.findContours (src, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
dst = cv2.cvtColor(src, cv2.COLOR_GRAY2BGR)
idx = 0

while idx >=0:
    c = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
    cv2.drawContours (dst, contours, idx, c, 2, cv2.LINE_8, hier)
    idx = hier [0, idx, 0]

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
